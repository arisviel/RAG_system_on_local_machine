```python
import pickle
from sentence_transformers import SentenceTransformer
from scipy import spatial



with open("./data/corpus.pickle", "rb") as handle:
    corpus = pickle.load(handle)
with open("./data/embeds.pickle", "rb") as handle:
    embeds = pickle.load(handle)
model = SentenceTransformer("intfloat/multilingual-e5-small")


```

```python
prompt = model.encode('Какими качествами обладают керамические фильтры при очистки высоко температурных газов')
sim_dict = {}

for emb_key in embeds.keys():
    sim = 1 - spatial.distance.cosine(prompt, embeds[emb_key])
    sim_dict[emb_key] = sim
sim_dict = dict(reversed(sorted(sim_dict.items(), key=lambda item: item[1])))
for ks in list(sim_dict.keys())[:3]:
    print(ks)
    print(corpus[ks])
```
